# TRANSACTION TIME PATTERN ANALYSIS

```python
# Convert 'Time' into hours
df['Hour'] = (df['Time'] // 3600) % 24  # Convert seconds to hours
```

```python
# --- Time Segmentation ---

# Segment into day vs. night
df['TimeOfDay'] = df['Hour'].apply(lambda x: 'Day' if 6 <= x < 18 else 'Night')

# Segment into business vs. non-business hours (assuming business hours are 9 AM to 5 PM)
df['BusinessHours'] = df['Hour'].apply(lambda x: 'Business Hours' if 9 <= x < 17 else 'Non-Business Hours')
```

```python
# --- Visualization ---

# 1. Transaction Count by Hour for Fraud and Non-Fraud Cases
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Hour', hue='Class', palette=['#3498db', '#e74c3c'])
plt.title('Transaction Count by Hour (Fraud vs. Non-Fraud)')
plt.xlabel('Hour of the Day')
plt.ylabel('Transaction Count')
plt.legend(['Non-Fraud', 'Fraud'])
plt.show()
```

```python
# 2. Day vs. Night Transaction Frequency for Fraud and Non-Fraud Cases
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='TimeOfDay', hue='Class', palette=['#3498db', '#e74c3c'])
plt.title('Day vs. Night Transaction Count (Fraud vs. Non-Fraud)')
plt.xlabel('Time of Day')
plt.ylabel('Transaction Count')
plt.legend(['Non-Fraud', 'Fraud'])
plt.show()
```

```python
# 3. Business Hours vs. Non-Business Hours Transaction Frequency for Fraud and Non-Fraud Cases
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='BusinessHours', hue='Class', palette=['#3498db', '#e74c3c'])
plt.title('Business vs. Non-Business Hours Transaction Count (Fraud vs. Non-Fraud)')
plt.xlabel('Time Segment')
plt.ylabel('Transaction Count')
plt.legend(['Non-Fraud', 'Fraud'])
plt.show()
```

```python
# 4. Time-based Heatmap for Fraud and Non-Fraud Transactions
# Create a pivot table for a heatmap with hours and fraud indicator
hourly_fraud_counts = df.pivot_table(index='Hour', columns='Class', values='Amount', aggfunc='count').fillna(0)
hourly_fraud_counts.columns = ['Non-Fraud', 'Fraud']

plt.figure(figsize=(12, 6))
sns.heatmap(hourly_fraud_counts, annot=True, fmt=".0f", cmap="Blues", linewidths=0.5)
plt.title('Hourly Fraud and Non-Fraud Transaction Heatmap')
plt.xlabel('Transaction Type')
plt.ylabel('Hour of the Day')
plt.show()
```

```python
# Class Distribution Analysis
class_counts = df['Class'].value_counts()
total_transactions = class_counts.sum()
fraud_percentage = (class_counts[1] / total_transactions) * 100
non_fraud_percentage = (class_counts[0] / total_transactions) * 100

print("Class Distribution Analysis:")
print(f"Total Transactions: {total_transactions}")
print(f"Fraudulent Transactions: {class_counts[1]} ({fraud_percentage:.2f}%)")
print(f"Non-Fraudulent Transactions: {class_counts[0]} ({non_fraud_percentage:.2f}%)\n")
```

```python
# Descriptive Statistics by Class
stats_class = df.groupby('Class')[['Time', 'Amount']].agg(['mean', 'median', 'std'])
print("Descriptive Statistics for 'Time' and 'Amount' by Class:")
print(stats_class, "\n")
```

```python
# Transaction Count by Hour for Fraud and Non-Fraud Cases
hourly_counts = df.groupby(['Hour', 'Class']).size().unstack(fill_value=0)
print("Transaction Count by Hour (Fraud vs. Non-Fraud):")
print(hourly_counts, "\n")
```

```python
# Day vs. Night Transaction Frequency for Fraud and Non-Fraud Cases
time_of_day_counts = df.groupby(['TimeOfDay', 'Class']).size().unstack(fill_value=0)
print("Day vs. Night Transaction Count (Fraud vs. Non-Fraud):")
print(time_of_day_counts, "\n")
```

```python
# Business Hours vs. Non-Business Hours Transaction Frequency for Fraud and Non-Fraud Cases
business_hours_counts = df.groupby(['BusinessHours', 'Class']).size().unstack(fill_value=0)
print("Business vs. Non-Business Hours Transaction Count (Fraud vs. Non-Fraud):")
print(business_hours_counts, "\n")
```

```python
# Hourly Fraud and Non-Fraud Transaction Counts
hourly_fraud_counts = df.pivot_table(index='Hour', columns='Class', values='Amount', aggfunc='count').fillna(0)
hourly_fraud_counts.columns = ['Non-Fraud', 'Fraud']
print("Hourly Fraud and Non-Fraud Transaction Counts:")
print(hourly_fraud_counts)
```

# TRANSACTION AMOUNT ANALYSIS

```python
# Amount Segmentation: Create bins for transaction amounts
bins = [0, 50, 100, 200, 500, 1000, 5000]  # Define the bins
labels = ['Close to None', 'Very Low', 'Low', 'Medium', 'High', 'Very High']  # Define labels for the bins
df['AmountCategory'] = pd.cut(df['Amount'], bins=bins, labels=labels, right=False)  # Adjust labels to match bins
```

```python
# Calculate fraud rates within each amount bin
fraud_rates = df.groupby('AmountCategory')['Class'].mean() * 100  # Mean gives the fraud rate in percentage

print("Fraud Rates within Each Amount Bin:")
print(fraud_rates)
```

```python
# Visualization: Plot fraud and non-fraud distributions within each amount bin

# Box Plot: Amount distribution by Class within each bin
plt.figure(figsize=(12, 6))
sns.boxplot(x='AmountCategory', y='Amount', hue='Class', data=df, palette=['#3498db', '#e74c3c'])
plt.title('Transaction Amount Distribution by Class within Each Amount Bin')
plt.xlabel('Amount Category')
plt.ylabel('Transaction Amount')
plt.legend(title='Class', labels=['Non-Fraud', 'Fraud'])
plt.show()
```

```python
# Bar Chart: Count of Fraud vs Non-Fraud Transactions in Each Amount Bin
amount_bin_counts = df.groupby(['AmountCategory', 'Class']).size().unstack(fill_value=0)
amount_bin_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['#3498db', '#e74c3c'])
plt.title('Count of Fraud vs Non-Fraud Transactions in Each Amount Bin')
plt.xlabel('Amount Category')
plt.ylabel('Count of Transactions')
plt.xticks(rotation=45)
plt.legend(title='Class', labels=['Non-Fraud', 'Fraud'])
plt.show()
```

```python
# Violin Plot: Distribution of Amounts for Fraud and Non-Fraud Transactions
plt.figure(figsize=(12, 6))
sns.violinplot(x='AmountCategory', y='Amount', hue='Class', data=df, split=True, palette=['#3498db', '#e74c3c'])
plt.title('Violin Plot of Transaction Amounts by Class within Each Amount Bin')
plt.xlabel('Amount Category')
plt.ylabel('Transaction Amount')
plt.legend(title='Class', labels=['Non-Fraud', 'Fraud'])
plt.show()
```

```python
# Additional Analysis: Descriptive statistics for each amount category by Class
amount_category_stats = df.groupby(['AmountCategory', 'Class'])['Amount'].agg(['mean', 'median', 'std', 'count']).reset_index()
print("Descriptive Statistics for Transaction Amounts by Class and Amount Category:")
print(amount_category_stats)
```

# CROSS_ANALYSIS OF TIME AND AMOUNT

```python
# Create bins for transaction time (in hours)
bins_time = [0, 6*3600, 12*3600, 18*3600, 24*3600]  # Bins for [0-6), [6-12), [12-18), [18-24)
labels_time = ['Night', 'Morning', 'Afternoon', 'Evening']
df['TimeCategory'] = pd.cut(df['Time'], bins=bins_time, labels=labels_time, right=False)
```

```python
# Cross-tabulate Time intervals with Amount bins
cross_tab = pd.crosstab(df['TimeCategory'], df['AmountCategory'], values=df['Class'], aggfunc='sum').fillna(0)
cross_tab.columns.name = 'Amount Category'
cross_tab.index.name = 'Time Category'

print("Cross-tabulation of Time and Amount Categories (Fraud Counts):")
print(cross_tab)
```

```python
# Conditional Probability Analysis: Calculate probability of fraud for each combination
total_fraud_counts = df['Class'].sum()
probability_fraud = cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100  # Convert to percentage

print("\nConditional Probability of Fraud for Time and Amount Combinations (%):")
print(probability_fraud)
```

```python
# Visualization: Heatmap of Fraud Probability across Time and Amount Combinations
plt.figure(figsize=(10, 6))
sns.heatmap(probability_fraud, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'Probability of Fraud (%)'})
plt.title('Heatmap of Fraud Probability by Time and Amount Categories')
plt.xlabel('Amount Category')
plt.ylabel('Time Category')
plt.show()
```

```python
# Grouped Bar Chart: Show fraud counts across different time and amount combinations
fraud_counts = cross_tab.copy()
fraud_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['#3498db', '#e74c3c'])
plt.title('Fraud Counts by Time and Amount Categories')
plt.xlabel('Time Category')
plt.ylabel('Count of Fraudulent Transactions')
plt.xticks(rotation=45)
plt.legend(title='Amount Category')
plt.show()
```

```python
# Additional Analysis: Descriptive statistics for fraud by Time and Amount categories
fraud_stats = df[df['Class'] == 1].groupby(['TimeCategory', 'AmountCategory'])['Amount'].agg(['mean', 'median', 'std', 'count']).reset_index()
print("\nDescriptive Statistics for Fraudulent Transactions by Time and Amount Categories:")
print(fraud_stats)
```
